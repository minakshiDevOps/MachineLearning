```python
# Creating a sample automobile dataset with some incorrect data
data = {
    'Manufacturer': ['Ford', 'Toyota', 'Honda', 'Ford', 'Toyota'],
    'Model': ['F-150', 'Camry', 'Civic', 'Mustang', 'Corolla'],
    'Type': ['Truck', 'Sedan', 'Sedan', 'Coupe', 'Sedan'],
    'Min.Price': ['20000', 24000, 22000, 'N/A', 21000],  # 'N/A' as a string
    'Price': [30000, '27000', 'N/A', 31000, 25000],  # Incorrect format in 'Price' column
    'Max.Price': [35000, 30000, 'N/A', 36000, '28000'],  # 'N/A' and strings for numbers
    'MPG.city': [20, 28, 30, 'N/A', 27],  # 'N/A' in MPG.city
    'Cylinders': [6, 4, 'Four', 8, 'Four'],  # Mixed types (numbers and words)
    'Horsepower': [250, '200', 'One-forty', 300, 150],  # Mixed numeric and text data
    'Weight': [4500, 3200, 2900, 'Unknown', 3000]  # 'Unknown' as incorrect data
}

# Creating DataFrame
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
# To clean up numeric columns containing non-numeric values, 
# we can various functions to attempt type conversion and handle errors

# Correcting Data Formats:- 
# Can be done using to_numeric() or astype() for Data Type Conversion
```

```python
# extra note for error
```

```python
# Convert 'Price', 'Min.Price', and 'Max.Price' to numeric, forcing errors to NaN

df['Price'] = pd.to_numeric(df['Price'], errors = 'coerce')
df['Min.Price'] = pd.to_numeric(df['Min.Price'], errors = 'coerce')
df['Max.Price'] = pd.to_numeric(df['Max.Price'], errors = 'coerce')

df.dtypes
```

```python
df
```

```python
# Clean 'Cylinders' column by replacing text values and converting to int

# Using chain function - df['col'].replace({..}).astype(..)

df['Cylinders'] = df['Cylinders'].replace({'Four': 4}).astype(float, errors='ignore')
df
```

```python
# Handling Date and Time with pd.to_datetime()
# If your dataset contains date columns stored as strings, 
# you can use pd.to_datetime() to standardize them

# Sample data with a Date column
df['SaleDate'] = ['2023-01-15', '2022-12-25', '2022-11-10', '2023-01-05', '2022-10-30']

df['SaleDate'] = pd.to_datetime(df['SaleDate'])
df
```

```python
# Apply a function to create new columns based on existing data, 
# such as categorizing cars based on price

# Create a new column 'Price_Category' based on 'Price'
'''
Price              Category
< 25000           'Budget'
price < 35000     'Midrange'
others            'Luxury'
'''

def price_category(price):
    if price<25000:
        return 'Budget'
    elif price < 35000:
        return 'Midrange'
    else:
        return 'Luxury'


df['Price_Category'] = df['Price'].apply(price_category)

df[['Manufacturer','Model','Price', 'Price_Category']]
```
